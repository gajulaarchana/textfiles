```python
#Find	City wise list all the Villa which is not less than ten thousand.
filtered_villas = data[(data['type'] == 'villa') & (data['price'] >= 10000)]
city_villa_groups = filtered_villas.groupby('city')
for city, villas in city_villa_groups:
    print(f"City: {city}")
    for _, villa in villas.iterrows():
        print(f"Villa: {villa['street']} - Price: {villa['price']}")
    print("=" * 30)
```

```python
#In GALT city which residential type has more than 800sq__ft. Display their respective details street,sq__ft,sale_date,city.
filtered_properties = data[(data['city'] == 'GALT') & (data['type'] == 'Residential') & (data['sq__ft'] > 800)]
for index, property_details in filtered_properties.iterrows():
    print("Street:", property_details['street'])
    print("Square Feet:", property_details['sq__ft'])
    print("Sale Date:", property_details['sale_date'])
    print("City:", property_details['city'])
```

```python
#Which is the cheapest Villa in CA. name the city,street and price for the Villa
cheapest_villa = data[(data['state'] == 'CA') & (data['type'] == 'villa')].sort_values(by='price').iloc[0]
print("City:", cheapest_villa['city'])
print("Street:", cheapest_villa['street'])
print("Price:", cheapest_villa['price'])
```

```python
#List top 5 residency details which lie in the budget of 60000-120000, an area more than 1450, min bedroom 3 and, min bathroom 2.
top_5_properties = data.copy()
filtered_properties = top_5_properties.query(
    "60000 <= price <= 120000 and sq__ft > 1450 and beds >= 3 and baths >= 2"
)
print(filtered_properties.head(5))
```
